```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
reshape = np.reshape #lo que hace es que si tienes un arreglo de 1x100 lo convierte en 10x10, pasandolo a 2 dimensiones
```

```python
x=np.radians(np.arange(720))#crea un arreglo de 720 valores de 0 a 360 grados
plt.figure()#crea una figura

#Primer gráfico
plt.subplot(2,1,1)#2 filas, 1 columna, 1 gráfico
plt.plot(x,np.sin(x),label='seno',color="green")#grafica seno

#Agregar segundo gráfico
plt.subplot(2,1,2)#2 filas, 1 columna, 2 gráfico
plt.plot(x,np.cos(x),label='coseno',color="yellow")#grafica coseno
plt.show()
```

```python
plt.style.use('seaborn-whitegrid')#estilo de gráfica
fig=plt.figure()#crea una figura
ax=plt.axes()#crea un eje
x=np.linspace(0,10,1000)#crea un arreglo de 1000 valores de 0 a 10
ax.plot(x,np.sin(x),linestyle="-.")#grafica seno
plt.show()
```
